```python
%matplotlib inline

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

dc_data = "/Users/datki/Desktop/DC_data/DC_Properties.csv"
```

```python
df = pd.read_csv(dc_data)
df.head()
```

```python
isolated_pd = df.drop(columns=["Unnamed: 0", "BATHRM", "HF_BATHRM", "HEAT", "AC", "NUM_UNITS", 
                                                "ROOMS", "BEDRM", "AYB", "YR_RMDL", "EYB", "STORIES", "SALE_NUM", 
                                                "BLDG_NUM", "GRADE", "CNDTN", "EXTWALL", "ROOF", "INTWALL",
                                                "KITCHENS", "FIREPLACES", "USECODE", "GIS_LAST_MOD_DTTM",
                                                "CMPLX_NUM", "ASSESSMENT_NBHD", "ASSESSMENT_SUBNBHD", 
                                                "CENSUS_TRACT", "CENSUS_BLOCK", "QUADRANT", "X", "Y"])
isolated_pd.head()
```

```python
final_df= df.loc[df['QUALIFIED']=='Q']
final_df.plot.scatter(x='GBA', y='PRICE')
```

```python
new_df = isolated_pd.fillna(0)
new_df.head()
```

```python
new_data = new_df[new_df.PRICE != 0]
new_data.head()
```

```python
year = new_data['SALEDATE']
price_data = new_data['PRICE']
plt.scatter(year, price_data, edgecolors='r')
plt.xlabel('SALEDATE')
plt.ylabel('PRICE')
plt.show()
```

```python
style_data = new_data["STYLE"]
price_data = new_data["PRICE"]
colors = ["#1f77b4", "#ff7f0e", "#2ca02c", "#d62728", "#8c564b"]
explode = (0.1, 0, 0, 0, 0)  
plt.pie(price_data, labels=style_data, explode=None, colors=None,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.axis("equal")
# Save an image of our chart and print the final product to the screen
plt.savefig("../PyPies.png")
plt.show()
```
